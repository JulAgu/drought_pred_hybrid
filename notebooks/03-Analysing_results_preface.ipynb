## Just a fast test to imagine a map format for the results imitating Minixhofer et al.

```python
import geopandas as gpd
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
usa_map = gpd.read_file("../src/counties.geojson")
```

```python
continental_usa_map = usa_map.cx[-130:-60, 24:50]
```

```python
continental_usa_map.head()
```

```python
continental_usa_map.loc[:, "fips"] = continental_usa_map.loc[:, "STATEFP"] + continental_usa_map.loc[:, "COUNTYFP"]
```

```python
continental_usa_map = continental_usa_map.astype({"fips": "int32"})
```

```python
continental_usa_map.plot()
```

```python
soil_data = pd.read_csv("../src/soil_data.csv")
```

```python
continental_usa_map
```

```python
soil_data
```

```python
new_map = pd.merge(continental_usa_map, soil_data.loc[:, ["fips", "SQ7"]], on="fips")
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 5))
cax = new_map.plot(column="SQ7", cmap=matplotlib.colormaps.get_cmap('Set1').resampled(7), ax=ax, legend=True)
plt.show()
```
